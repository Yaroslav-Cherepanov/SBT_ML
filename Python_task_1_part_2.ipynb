Разберемся что такое linspace

```python
print np.linspace(-100, 100)
print len(np.linspace(-100, 100))
```

```python
def my_linspace(a, b, num=50):
    point_dist = (b-a)/float(num-1)
    return [a+point_dist*i for i in xrange(num)]
```

```python
start = -1000
stop = 1000
num = 50
print np.all(my_linspace(start, stop, num) == np.linspace(start, stop, num))

start = -10000
stop = 10000
num = 75
print np.all(my_linspace(start, stop, num) == np.linspace(start, stop, num))

# (вообще это можно было засунуть в doctest)
```

```python
%timeit np.linspace(int(-1e6), int(1e6))
```

```python
%timeit my_linspace(int(-1e6), int(1e6))
```

Кажется что наша версия linspace довольно неплохая, но на самом деле, если увеличить число отрезков, на которые дробить, то можно доказать обратное:

```python
%timeit my_linspace(int(-1e6), int(1e6), num=10000)
```

```python
%timeit np.linspace(int(-1e6), int(1e6), num=10000)
```

Конечно, это обьясняется не слишком эффективной реализацией: в моей функции на каждой итерации происходит умножение, хотя можно просто прибавлять к предыдущему длину отрезка. Но сейчас перед нами не стоит задачи эффективной реализации.

Нарисуем heatmap функции $ f(x, y) = 3xy + x − 2y $:

```python
x_ax = y_ax = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x_ax, y_ax)
Z = 3*X*Y + X - 2*Y

plt.pcolormesh(x_ax, y_ax, Z, cmap = plt.cm.coolwarm)
```

Теперь та же функция в 3D:

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 5 Всё вместе + SciPy

```python
import scipy.stats as sts
import pandas as pd

x = np.linspace(-4, 4, 500)
mu = 0
sigma = 3
noise = sts.norm(loc=mu, scale=sigma).rvs(size=500)

def f(x):
    return x**3 + 2*x*x - 3*x + 2

y = f(x)
y_noised = y + noise

pd.Series(data=y, index=x).plot(color='r')
plt.plot(x, y_noised, 'o')
```

```python
def f_opt(w):
    return sum((w[0]*x**3 + w[1]*x*x + w[2]*x + w[3] - y_noised)**2)

from scipy.optimize import minimize
w = minimize(f_opt, [0,0,0,0]).x

def f_cap(w,x):
    return w[0]*x**3 + w[1]*x*x + w[2]*x + w[3]
```

```python
from scipy.signal import argrelextrema
pd.Series(data=y, index=x).plot(color='r')
y_cap = f_cap(w, x)
pd.Series(data=y_cap, index=x).plot(color='b')

plt.axhline(y_cap[argrelextrema(y_cap, np.greater)], color='y')
plt.axhline(y_cap[argrelextrema(y_cap, np.less)], color='y')
```

Видно что графики практически совпадают

Можно решить и СЛАУ:

```python
four_x_points = [1, -2, .5, -3]
A = []
y = []
for curr_x in four_x_points:
    A.append([curr_x**3, curr_x**2, curr_x**1, curr_x**0])
    y.append(f(curr_x))
```

```python
print A
print y
```

Получаем верные коэффициенты многочлена:

```python
np.linalg.solve(A, y)
```

## 6 Pandas + Scikit-learn

```python
init_data = pd.read_csv('Skin_NonSkin.txt', sep='\t', header=None)
init_data.columns = ['B', 'G', 'R', 'target']
train_init = init_data[['B', 'G', 'R']]
y = init_data['target'] - 1
# train_full = train_init.copy()
train_changed = pd.DataFrame()
train_changed['B_mul_G'] = train_init.B * train_init.G
train_changed['B_mul_R'] = train_init.B * train_init.R
train_changed['G_mul_R'] = train_init.G * train_init.R
train_changed['B_diff_G'] = abs(train_init.B - train_init.G)
train_changed['B_diff_R'] = abs(train_init.B - train_init.R)
train_changed['G_diff_R'] = abs(train_init.G - train_init.R)
```
